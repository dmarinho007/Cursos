```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
sns.set_palette('rocket')
sns.set_style('darkgrid')
ax = sns.lineplot(x ='mes', y ='vendas', data = alucar)
ax.figure.set_size_inches(16,8)
ax.set_title('Vendas Alucar de 2017/2018' , loc='left', fontsize=18)
ax.set_xlabel('Tempo', fontsize = 16)
ax.set_ylabel('Vendas', fontsize = 16)
ax = ax
```

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar.head(5)
```

```python
sns.set_palette('rocket')
sns.set_style('darkgrid')
ax = sns.lineplot(x ='aumento', y ='vendas', data = alucar)
ax.figure.set_size_inches(16,8)
ax.set_title('Aumento das vendas Alucar de 2017/2018' , loc='left', fontsize=18)
ax.set_xlabel('Tempo', fontsize = 16)
ax.set_ylabel('Aumento', fontsize = 16)
ax = ax
```

```python
def plotar (titulo, labelx, labely, x, y, dataset):
    sns.set_palette('rocket')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x = x, y = y, data = dataset)
    ax.figure.set_size_inches(16,8)
    ax.set_title(titulo , loc='left', fontsize=18)
    ax.set_xlabel(labelx, fontsize = 16)
    ax.set_ylabel(labely, fontsize = 16)
    ax = ax
```

```python
plotar('Aumento das vendas Alucar de 2017/2018', 'Tempo', 'Aumento', 'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Acelerao das vendas Alucar 2017/2018', 'Tempo', 'Acelerao', 'mes', 'aceleracao', alucar )
```

```python
alucar.sample(10)
```

```python
plt.figure(figsize = (18,14))
ax = plt.subplot(3, 1, 1)
ax.set_title('Anlise de vendas Alucar 2017/2018', fontsize = 18)
ax = sns.lineplot(x ='mes', y ='vendas', data = alucar)
ax = plt.subplot(3, 1, 2)
ax = sns.lineplot(x ='mes', y ='aumento', data = alucar)
ax = plt.subplot(3, 1, 3)
ax = sns.lineplot(x ='mes', y ='aceleracao', data = alucar)
ax = ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize = (18,14))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 18)
    ax = sns.lineplot(x = x, y = y1, data = dataset)
    ax = plt.subplot(3, 1, 2)
    ax = sns.lineplot(x = x, y = y2, data = dataset)
    ax = plt.subplot(3, 1, 3)
    ax = sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
    
```

```python
plot_comparacao('mes','vendas','aumento','aceleracao'
            , alucar, 'Anlise de vendas Alucar 2017/2018')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao das vendas', fontsize = 18)
autocorrelation_plot(alucar['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao do aumento', fontsize = 18)
autocorrelation_plot(alucar['aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao da acelerao', fontsize = 18)
autocorrelation_plot(alucar['aceleracao'][2:])
ax = ax
```

```python
vendas_diarias['aumento'] = vendas_diarias['vendas'].diff()
vendas_diarias['aceleracao'] = vendas_diarias['aumento'].diff()
vendas_diarias.head()
```

```python
plot_comparacao('dia','vendas','aumento','aceleracao',vendas_diarias, 'Anlise de vendas Out/Nov - Chocolura')
```

```python
vendas_diarias['dia_semana'] = vendas_diarias['dia'].dt.day_name()
```

```python
vendas_diarias['dia_semana'].unique()
```

```python
dias_traduzidos = {'Monday' : 'Segunda', 
                   'Tuesday' : 'Tera', 
                   'Wednesday' : 'Quarta', 
                   'Thursday' : 'Quinta', 
                   'Friday' : 'Sexta', 
                   'Saturday' : 'Sbado',
                   'Sunday' : 'Domingo'
    
}
```

```python
vendas_diarias['dia_semana'] = vendas_diarias['dia_semana'].map(dias_traduzidos)
```

```python
vendas_diarias.head()
```

```python
vendas_agrupadas = vendas_diarias.groupby('dia_semana')['vendas','aumento','aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao das vendas dirias', fontsize = 18)
autocorrelation_plot(vendas_diarias['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao do aumento das vendas dirias', fontsize = 18)
autocorrelation_plot(vendas_diarias['aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize = (16,10))
ax.suptitle('Correlao da acelerao de vendas dirias', fontsize = 18)
autocorrelation_plot(vendas_diarias['aceleracao'][2:])
ax = ax
```

#            # Statsmodel

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolura['vendas'], freq = 3) 
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = {
    'observacao':observacao,
    'tendencia':tendencia,
    'sazonalidade':sazonalidade,
    'ruido':ruido
}

resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(resultado.index, 'observacao','tendencia','sazonalidade',resultado,'Exemplo de statsmodels')
```
