# Estatistica

```python
dataset = np.column_stack((anos, km, valor))
```

```python
dataset.shape
```

```python
np.mean(dataset, axis = 0)
```

```python
np.mean(dataset[:, 2])
```

```python
np.std(dataset[:, 2])
```

```python
dataset.sum(axis = 0)
```
